```python
df = pd.read_csv('2020 01-15 test timeseries.csv')
```

```python
dfg = df.groupby(['timestamp','person'])
dfg.aggregate(['mean','std']).unstack()
```

```python
fig, ax = plt.subplots(figsize=(6,4))
means = df.groupby(['timestamp','person']).mean()['value'].unstack()
errors = df.groupby(['timestamp','person']).std()['value'].unstack()

means.plot(yerr=errors,ax=ax)
```
