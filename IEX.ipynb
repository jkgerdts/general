```python

import requests

import pandas as pd
import io
import os
import json
os.chdir('/Users/jkgerdts/')
with open('config.json') as config_file:
    data = json.load(config_file)
    os.environ['TOKEN'] = data['TOKEN']

def get_csv(api_url):
    
    with requests.Session() as session:
        response = session.get(api_url)
        raw_data = response.content.decode('utf8')
        data = pd.read_csv(io.StringIO(raw_data), sep = ',')
    return(data)

def format_url(sym,datestr,iexonly,sandbox):
    api_url = 'https://{}.iexapis.com/stable/stock/{}/chart/date/{}?format=csv&token={}{}'
    if sandbox:
        site = 'sandbox'
    else:
        site = 'cloud'
    if iexonly:
        iex='&chartIEXOnly=true'
    else:
        iex=''
    import json
    api_url = api_url.format(site,sym,datestr,os.environ['TOKEN'],iex)
    return api_url

def get_data(sym):
    from datetime import datetime, timedelta
    import time
    datestrings = []
    for i in range(30):
        t = datetime.now()-timedelta(days=i)
        if t.weekday() < 5:
            #print(t.weekday(),t.strftime('%Y%m%d'))
            datestrings.append(t.strftime('%Y%m%d'))

    os.chdir('/Users/jkgerdts/stock/')
    iexonly = True # if true accesses only the free data on IEX cloud excluding larger market data
    sandbox = False # if true accesses only fake free data

    for i in range(len(datestrings)):
        if sym+'_'+datestrings[i]+'.csv' not in os.listdir():
            data = get_csv(format_url(sym,datestrings[i],iexonly,sandbox))
            data.to_csv(sym+'_'+datestrings[i]+'.csv')
        else:
            print('file '+sym+'_'+datestrings[i]+'.csv already exists')
    return None
```

```python
syms = ['zgnx','aapl','goog','vrtx','ions','ndaq','biib','frgi','ibb','fivn','blue','gpro','zyne','tsla']
for sym in syms:
    get_data(sym)
```
