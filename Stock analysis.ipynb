```python
def getdata(sym,announcedate):
    import yfinance as yf
    from datetime import datetime as dt
    from datetime import timedelta
    #from numpy.random import shuffle
    tformat = '%Y-%m-%d'
    
    t_announce = dt.strptime(announcedate,tformat)
    t_start=t_announce-timedelta(days=30)
    t_end=t_announce+timedelta(days=30)

    data = yf.download(tickers=sym,start=t_start.strftime(tformat),end=t_end.strftime(tformat),interval="1d")

    vals = data['Open'].values/data['Open'].values[0]

    return vals



dates = '2019-08-16 2019-04-23 2019-12-23 2019-11-20 2019-06-21 2019-04-09 2019-09-12 2019-12-20 2019-07-30 2019-11-14 2019-08-16 2019-11-08 2019-10-08 2019-08-02 2019-02-01 2019-12-20 2019-11-25 2019-12-20 2019-03-20 2019-04-12 2019-10-11 2019-07-16 2019-11-15 2019-10-07 2019-02-13 2019-03-26 2019-05-24 2019-05-03 2019-11-14 2019-12-12 2019-07-25 2019-10-21'.split(' ')
symbols = 'ABBV ABBV AGN ALNY AMAG AMGN ARDX AZN BAYRY BGNE CELG CELG CLVLY DSNKY EOLS ESALY GBT ITCI JAZZ JNJ LLY MRK NVS NVS NVS NVS NVS PFE SGIOY SRPT STX VRTX'.split(' ')



###################### INCLUDE THIS BLOCK TO MAKE THE DATA CONTROL DATA
shuffle(dates)
##################### 

import pandas as pd
mydf = pd.DataFrame()

for d,s in zip(dates,symbols):
    try:
        mydf[s]=pd.Series(data=getdata(s,d))
    except ValueError:
        print('value error for'+s)
        pass
```
