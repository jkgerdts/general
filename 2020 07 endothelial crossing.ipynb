```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
ds = pd.read_clipboard()
#/Users/jkgerdts/Google Drive/LAB/data/2020 07-05 transwell pilots/2020 07-05 cxcl12 gradients.xlsx
ds=ds[ds['use']==1]
```

```python
grouped = ds.groupby(['huvecs','cxcl12'])
print(grouped['migrated'].mean())
```

```python
fig, ax = plt.subplots(figsize=(6,4))
means = ds[ds['huvecs']=='yes'].groupby(['cxcl12']).mean()['migrated']
errors = ds[ds['huvecs']=='yes'].groupby(['cxcl12']).std()['migrated']

means.plot(yerr=errors,ax=ax)
plt.xscale('log')
plt.title('# migrated through HUVECs (estimated)')
```

```python
fig, ax = plt.subplots(figsize=(6,4))
means = ds[ds['huvecs']=='no'].groupby(['cxcl12']).mean()['migrated']

means.plot(ax=ax)
plt.xscale('log')
plt.title('# migrated in absence of HUVECs')
```
